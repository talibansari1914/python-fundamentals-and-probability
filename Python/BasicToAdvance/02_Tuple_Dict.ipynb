# Dict comprehension

```python
flavors = ["vanilla", "chocolate", "strawberry", "mango", "coffee"]
prices_per_scoop = {flavor: len(flavor) * 5 for flavor in flavors}  # Simple logic: price = length of name  5
print(f"Ice cream prices: {prices_per_scoop}")
```

```python
premium_flavors = {flavor: price for flavor, price in prices_per_scoop.items() if price > 30}
print(f"Premium flavors: {premium_flavors}")
```

```python
fruits=["banana","apple"]
price_per_kg={x:len(fruits)*7 for x in fruits}
print(f"prices of the fruits:{price_per_kg}")
```

```python
list_bikes=["tvs","bmw","honda","royal enfield"]
quantity={quant: quant*2 for quant in list_bikes}
print(f"quantity of bikes are:{quantity}")
```

```python
list_bikes=["tvs","bmw","honda","royal enfield"]
quantity={quant: list_bikes*2 for quant in list_bikes}
print(f"quantity of bikes are:{quantity}")
```

```python
list_bikes=["tvs","bmw","honda","royal enfield"]
quantity={list_bike: len(list_bike)*2 for list_bike in list_bikes}
print(f"quantity of bikes are:{quantity}")
```
