```python
def str_name(s):
    rev_str=" "
    for i in range(len(s)-1,-1,-1):
        rev_str+=s[i]
    return rev_str
```

# Set functions:-
1.set_name.add()     --> it takes one arg and this to the set
2.set_name.remove()  -->it takes the element name to remove it
3.set_name.pop()     -->it removes the element and return to variable in which it has stored
4.set_name.discard() -->it also removes the element which is given to it if the element does not found in the set then id does not generate "error"
5.set_name.clear()   -->it used to remove all the element from the set or to empty the set
6.set_name.copy()    -->it used to maintain the local copy of a set which is stored in the other variable
7.len(set_name)      -->it used to find the length of the set or how many elements a set have
## Set Operations
1.union-->set_a.union(set_b)--->it joins the two sets
2.intersection--> set_a.intersection(set_b)  --->it finds common in both sets
3.difference-->   set_a.difference(set_b)    --->it removes the common part from the set_a and then print it
4.symmetric--->   set_a^set_b                --->it removes the common part form both the sets and then add the sets
5.subset-->       set_a.issubset(set_b)      --->if set_a have only some elements of set_b and no other extra elements then it is print 'True' else 'false'
6.disjoint-->     set_a.isdisjoint(set_b)    ---> no elements will be equal from both sets or both set shuold contain the different-different element
