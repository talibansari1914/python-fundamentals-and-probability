# List 
1.list_name.append()---elements can be added in empty list,list,nested list
2.list_name.extend()---it iterates on the given exactly one argument and then add it into the list
3.list_name.insert(,)--it takes exactly 2 arguments one is index second one is what to insert in the list
4.list_name.index()----it takes one element as argument from the list and then return its index & does not provide the index of nested list element
5.list_name.remove()---give value or item from the list to remove it
6.list_name.pop()------give index of that value to remove it
7.list_name.sort()----it takes no arg & work for  integer & character both and for single list
8.list_name.reverse()-just reverse the list element not the character
9.list_name.count()---it takes one arg to count that number in the 1-D list not the nested list
