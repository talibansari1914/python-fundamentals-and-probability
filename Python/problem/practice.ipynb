```python
st="python is a coding langugage"
for i in st:
    if i=="coding".lower():
        i=i.capitalize()
        print(i)
```

```python
st="python is a coding langugage"
list_a=st.split()
for i,j in enumerate(list_a):
    if j=="coding".lower():
        list_a[i]=j.capitalize()
print(list_a)

```

```python
st1=" im a data science learner"
for i,j in enumerate(st1):
    if j in "dsl":
        st1[i]=j.capitalize()
        print(st1)
```

```python
st1=" im a data science learner"
list_b=st1.split()
for i,j in enumerate(list_b):
    if j=="d" or j=="s" or j=="l":
        list_b[i]=j.capitalize()
print(list_b)
```
