# Vector Operation : Combinantion & Projection
### 1.Linear Combonation of Vector

```python
# example 1.
import numpy as np
import matplotlib.pyplot as plt
# Recipe_Vector
R_A=np.array([8.0,3.0])
R_B=np.array([12.3,2.5])

# Quantity
Q_A=4
Q_B=3
scaled_A=Q_A*R_A
scaled_B=Q_B*R_B

total_requirement=scaled_A+scaled_B

print("Scaled Quantity(A):",scaled_A)
print("Scaled Quantity(B):",scaled_B)
print("Total Requirements:",total_requirement)

plt.figure(figsize=(9,9))
plt.quiver(0,0,R_A[0],R_A[1],angles='xy',scale_units='xy',scale=1,label='Original_A(8.0,3.0)',color='violet')
plt.quiver(0,0,R_B[0],R_B[1],angles='xy',scale_units='xy',scale=1,label='Original_B(12.3,2.5)',color='black')
plt.quiver(0,0,scaled_A[0],scaled_A[1],angles='xy',scale_units='xy',scale=1,label='Scaled A(32,12)',color='orange')
plt.quiver(0,0,scaled_B[0],scaled_B[1],angles='xy',scale_units='xy',scale=1,label='Scaled B(36.9,7.5)',color='blue')
plt.quiver(0,0,total_requirement[0],total_requirement[1],angles='xy',scale_units='xy',scale=1,label='Total Requirement(68.9,19.5)',color='red')

plt.xlim(0,70)
plt.ylim(0,70)
plt.xlabel("Quantity A",color='black',fontweight='bold',fontsize=10)
plt.ylabel("Quantity B",color='black',fontweight='bold',fontsize=10)
plt.title("Geometric Representation",fontweight='bold',fontsize=16,color="grey")
plt.grid(True,alpha=0.5)
plt.legend()
plt.show()
```

### Projection (Scaler&Vector)

```python
import numpy as np
# Price of the Product
A=np.array([1000.,1500.,700.])
#weights of product
B=np.array([2.,3.,1.])

# Dot Product A*B
dot_A_B=np.dot(A,B)

# dot product of B*B
dot_B=np.dot(B,B)

# Projection of the scaler
alpha=dot_A_B/dot_B

# Vector Projection
vector_projection=alpha*B

print("\nA:",A)
print("\nB:",B)
print("\nDot Product of A*B:",dot_A_B)
print("\nDot Product of B*B:",dot_B)
print("\nScale Projection or L2 Normalisation:",alpha)
print("\nVector Projection:",vector_projection)
```

```python
import torch
A_pt=torch.tensor([200.,300.,100.])
B_pt=torch.tensor([4.,3.,2.])

# dot product of A_pt*B_pt
dot_A_B_pt=torch.dot(A_pt,B_pt)
# dot product of B_pt*B_pt
dot_B_pt=torch.dot(B_pt,B_pt)

# L2 Normalisation or scaler Projection
scaler_projection=dot_A_B_pt/dot_B_pt

# Vector Projection
vector_projection=scaler_projection*B_pt

print("A_pt:",A_pt.tolist())
print("B_pt:",B_pt.tolist())
print("dot_A_B_pt:",dot_A_B_pt.item())
print("dot_B_pt:",dot_B_pt.item())
print("Scaler Projection or L2 Normalisation:",scaler_projection.item())
print("Vector Projection:",vector_projection.tolist())
```

```python
import numpy as np
import matplotlib.pyplot as plt
A_np=np.array([4,5])
B_np=np.array([6,4])

# Dot Product of A_np*B_np
dot_A_B_np=np.dot(A_np,B_np)
print(dot_A_B_np)
# Dot PRroduct of B_np*B_np
dot_B_np=np.dot(B_np,B_np)
print(dot_B_np)
# scaler PRojection
scale_np=dot_A_B_np/dot_B_np

# Vector Projection
vector_np=scale_np*B_np
print(vector_np)
plt.figure(figsize=(8,8))
plt.quiver(0,0,A_np[0],A_np[1],angles='xy',scale_units='xy',scale=1,label='Sales(A)',color='red')
plt.quiver(0,0,B_np[0],B_np[1],angles='xy',scale_units='xy',scale=1,label='Sales(B)',color='violet')
# Vector Projection
plt.quiver(0,0,vector_np[0],vector_np[1],angles='xy',scale_units='xy',scale=1,label='Vector(Sales)',color='green')
# plot a dotted line perform orthogonal projection
plt.plot([A_np[0],vector_np[0]],[A_np[1],vector_np[1]],'k--',label='orthogonal component',alpha=0.6)
plt.xlim(0,10)
plt.ylim(0,10)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title("Geometric Representation")
plt.grid(True,alpha=0.5)
plt.legend()
plt.show()
```
