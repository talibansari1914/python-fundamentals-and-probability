# Matrix Implementation
## Matrix Operation :

#### 1.Addition & Subtraction on Matrix

```python
import numpy as np
# sales Marix of ice-cream
a_sales=np.array([
    [200,304,590],
    [100,250,150],
    [1000,500,300]
])

b_sales=np.array([
    [600,500,400],
    [340,200,506],
    [200,866,900],
])

# matirx addition
total_sales=a_sales+b_sales

print("Sales of A:\n",a_sales)
print("Sales of B:\n",b_sales)
print("Total Sales:\n",total_sales)

print("====="*10)
reduced_sales=a_sales-b_sales
print("Reduced Sales:\n",reduced_sales)
```

```python
import torch
vipul_sales=torch.tensor([
    [150.,140.2,47.8],
    [474.3,484.,484.8],
    [384.3,584.47,595.48]
])

brijesh_sales=torch.tensor([
    [495.49,484,394.398],
    [953.39,430.39,283.38],
    [490.39,494.3,948.4]
])

# Total Sales
total_sales_pt=torch.add(vipul_sales,brijesh_sales)

# Reduced Sales
reduced_sales_pt=brijesh_sales-vipul_sales

print("Vipul sales:\n",vipul_sales)
print("Brijesh sales:\n",brijesh_sales)
print("Total sales:\n",total_sales_pt)
print("Reduced Sales:\n",reduced_sales_pt)
```

### 2.Scaler Multiplication

```python
import numpy as np
cost=np.array([2000,4738.38,2928.48])
gst=0.18

# increased Price after gst
gst_price=gst*cost
total_price=gst_price+cost

print("Cost:\n",cost)
print("Gst:\n",gst)
print("Gst_price:\n",gst_price)
print("Total Price with gst:\n",total_price)
```

```python
import torch
cost_pt=torch.tensor([50.25,100.50,150.75])
gst_pt=0.25
gst_price_pt=gst_pt*cost_pt
total_price_pt=gst_price_pt+cost_pt
print("Cost :\n",cost_pt.tolist())
print("Gst:\n",gst_pt)
print("Gst Price:\n",gst_price.tolist())
print("Total Price:\n",total_price_pt.tolist())
```

### 3. Matrix Multiplication

```python
import numpy as np
price=np.array([
    [100,150,200],
    [50.20,30.56,49.38],
    [200.46,495.4,80.48]
])

price_per_unit=np.array([
    [30.28],
    [50.39],
    [100.25]
])

# Matrix Multiplication
product=price@price_per_unit
product2=np.dot(price,price_per_unit)

print("Product1:\n",product)
print("Product2:\n",product2)
```

```python
import torch
price_pt=torch.tensor([
    [10.5,20.,30.]
]
)

quantity_pt=torch.tensor([
    [2.,3.,4.],
    [8.,9.,7.],
    [7.,5.,6.]
])

# Mulitplication
#prod1=torch.dot(price_pt,quantity_pt)
prod2=torch.matmul(price_pt,quantity_pt)
prod3=price_pt@quantity_pt
#print("Product 1:\n",prod1.tolist())
print("Product 2:\n",prod2.tolist())
print("Product 3:\n",prod3.tolist())
```

## Transpose

```python
import numpy as np
# product-kg & price
A_np=np.array([
    [10,200],
    [5,500],
    [4,350],
])

transpose1=A_np.T
transpose2=np.transpose(A_np)

print("Shape A_np:",A_np.shape)
print("A_np:\n",A_np)

print("Transpose1:\n",transpose1)
print("Transpose1 shape:",transpose1.shape)

print("Transpose2:\n",transpose2)
print("Transpose2 shape:",transpose2.shape)
```

### Transpose Operation
**Double Transpose**
$$(A^T)^T = A$$

```python
import numpy as np

# Product--Stock & Quantity
B_np=np.array([
    [20,30],
    [40,50],
    [100,200]
])

# transpose
transpose_B=B_np.T
print("Original Matrix Shape:",B_np.shape)
print("Original Matrix:\n",B_np)

print("Transpose1 Shape:",transpose_B.shape)
print("single Transpose:\n",transpose_B)

# Double Transpose
double_transpose=transpose_B.T
print("Double transpose shape:",double_transpose.shape)
print("Double transpose or original matrix:\n",double_transpose)
```

**sum of transpose**
$$(A + B)^T = A^T + B^T$$

```python
import numpy as np
C_np=np.array([
    [4,10],
    [8,20],
    [12,30]
])

D_np=np.array([
    [2,8],
    [4,12],
    [7,15]
])

# Single Transpose 
transpose_C=C_np.T
transpose_D=D_np.T

transpose_sum_C_D=transpose_C+transpose_D
C_D=C_np+D_np
transpose_C_D=C_D.T

print("shape transpose_c:",transpose_C.shape)
print("Transpose C:\n",transpose_C)
print("shape transpose_d:",transpose_D.shape)
print("Transpose D:\n",transpose_D)

print("Shape of sum of transpose:",transpose_sum_C_D.shape)
print("Sum of Transpose:\n",transpose_sum_C_D)

print("shape of sum Transpose C_D:\n",transpose_C_D.shape)
print("Sum of C_np+D_np then Transpose:\n",transpose_C_D)

print(transpose_sum_C_D is transpose_C_D)
print(transpose_sum_C_D== transpose_C_D)
```
