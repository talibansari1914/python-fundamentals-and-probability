# Orthoginality in Vector Spaces
### 1.Vectors with zero dot product

```python
import numpy as np
A=np.array([5.25,10.65])
B=np.array([-10.65,5.25])
dot_A_B=np.dot(A,B)
TOLERANCE=1e-9
is_orthogonal=abs(dot_A_B)<TOLERANCE
print("A:",A)
print("B:",B)
print("dot_A_B:",dot_A_B)
print(is_orthogonal)
```

```python
import torch
A_pt=torch.tensor([7.47,12.3])
B_pt=torch.tensor([-12.3,7.47])
dot_A_B_pt=torch.dot(A_pt,B_pt)
TOLERANCE=1e-9
is_orthogonal_pt=abs(dot_A_B_pt)<TOLERANCE

print("A:",A_pt)
print("B:",B_pt)
print("Dot_A_B_pt:",dot_A_B_pt)
print("is orthogonal?:",is_orthogonal_pt)
```

```python
import numpy as np
import matplotlib.pyplot as plt

i_cap=np.array([1.0,0.0])
j_cap=np.array([0.0,1.0])

# vector v to decompose
v=np.array([5,6])

i_v=v[0]*i_cap
j_v=v[1]*j_cap

plt.figure(figsize=(7,7))
plt.quiver(0,0,i_cap[0],i_cap[1],angles='xy',scale_units='xy',scale=1,color='yellow',label='i_cap(1,0)')
plt.quiver(0,0,j_cap[0],j_cap[1],angles='xy',scale_units='xy',scale=1,color='blue',label='j_cap(0,1)')
plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='red',label='decomposition vector(5,6)')
plt.quiver(0,0,i_v[0],i_v[1],angles='xy',scale_units='xy',scale=1,color='violet',label='i_v(5,0)')
plt.quiver(0,0,j_v[0],j_v[1],angles='xy',scale_units='xy',scale=1,color='black',label='i_v(0,6)')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('X-Composition of values',color='black')
plt.ylabel('Y-composition of values',color='black')
plt.title('Decomposition on orthonormal basis',color='green')
plt.grid(True,alpha=0.5)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
