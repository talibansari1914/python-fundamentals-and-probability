# Component Wise Addition & Subtraction

```python
import numpy as np
import matplotlib.pyplot as plt

stock_current=np.array([100,38,75])
stock_used=np.array([78,30,65])

stock_remaining=stock_current-stock_used

print(stock_remaining)
```

```python
import torch
stock_c=torch.tensor([100,38,75],dtype=torch.float32)
stock_u=torch.tensor([78,30,65],dtype=torch.float32)

stock_r=stock_c-stock_u
print("Remaining Stock:",stock_r.tolist())
```

# Tip to Tail Rule

```python
import numpy as np
import matplotlib.pyplot as plt
 # initial_pt
v_np=np.array([5,4])
# Tip 
u_np=np.array([2,3])
# resultant Displacement
r_np=v_np-u_np

print(r_np)

plt.figure(figsize=(7,7))
plt.quiver(0,0,v_np[0],v_np[1],angles='xy',scale_units='xy',scale=1,label='v_np(5,6)',color='gray')
plt.quiver(v_np[0],v_np[1],u_np[0],u_np[1],angles='xy',scale_units='xy',label='u_np(2,4)',color='violet')
plt.quiver(0,0,r_np[0],r_np[1],angles='xy',scale_units='xy',scale=1,label='Resultant(3,2)',color='black')

plt.xlim(0,7)
plt.ylim(0,7)
plt.xlabel('Initial Point (no displacement)')
plt.ylabel('Last Point(Displaced Point)')
plt.title('Geometric Represntation of Displacement')
plt.legend()
plt.grid(True)
plt.show()
```

# Properties of Vector Addition
## 1.Commutativity
**A+B=B+A**

```python
import numpy as np

a_np=np.array([28.46,48.23,29.49])
b_np=np.array([12.83,53.27,48.38])

result_ab=a_np+b_np
result_ba=b_np+a_np
print("a_np:",a_np)
print("b_np:",b_np)
print("result_ab:",result_ab)
print("result_ba:",result_ba)
print(result_ab is result_ba)
print(result_ab==result_ba)
```

## Additivi Identitiy
**A+0=A**

```python
import torch
a_t=torch.tensor([90.26,75.38,63.47],dtype=torch.float32)
#additive
result=a_t+0

print("a_t:",a_t)
print("result:",result)
print(a_t is result)
print(a_t == result)
```

## Inverse
**a+(-a)=0**

```python
import torch 
b_t=torch.tensor([5,4,8])
invers=-b_t

result_1=b_t+invers
print("b_t:",b_t)
print("inverse:",invers)
print("result_1:",result_1)
```
