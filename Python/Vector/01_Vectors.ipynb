## Vector Addition

### Numpy vector addition

```python
import numpy as np
#initial_pt-->starting location point
initial_pt=np.array([5,7])

# shift_pt-->shifted from starting point
shift_pt=np.array([3,2])

#new_pt-->new location point
new_pt=initial_pt+shift_pt

print("initial_pt:",initial_pt)
print("\nshift_pt:",shift_pt)
print("\new_pt:",new_pt)
```

```python
import torch
#initial_pt-->starting loaction point
initial_pt=torch.tensor([5,7])

#shift_pt-->shifted from starting point
shift_pt=torch.tensor([3,2])

#new_pt-->new location point
new_pt=initial_pt+shift_pt

print("initial_pt(pytorch):",initial_pt)
print("\nshift_pt(pytorch):",shift_pt)
print("\nnew_pt(pytorch):",new_pt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
#initial_pt-->starting location point
initial_pt=np.array([5,7])

# shift_pt-->shifted from starting point
shift_pt=np.array([3,2])

#new_pt-->new location point
new_pt=initial_pt+shift_pt

plt.figure(figsize=(7,7))
#draw initial_pt
plt.quiver(0,0,initial_pt[0],initial_pt[1],angles='xy',scale_units='xy',scale=1,color='black')

#draw shift_pt(from tip of initial_pt)
plt.quiver(initial_pt[0],initial_pt[1],shift_pt[0],shift_pt[1],angles='xy',scale_units='xy',scale=1,color='blue')

#draw new_pt=initial_pt+shift_pt
plt.quiver(0,0,new_pt[0],new_pt[1],angles='xy',scale_units='xy',scale=1,color='green',linestyle='dashed')

from matplotlib.lines import Line2D
legend_elements=[
    Line2D([0],[0],color='black',lw=2,label='Initial(5,7)'),
    Line2D([0],[0],color='blue',lw=2,label='Shift(3,2)'),
    Line2D([0],[0],color='green',lw=2,label='New Point(8,9)')
]

plt.xlim(-1,10)
plt.ylim(-1,10)
plt.grid(True)
plt.legend(handles=legend_elements)
plt.title('Geometric Vector Addition')
plt.xlabel('X-Coordinates(m)')
plt.ylabel('Y-Coordinates(m)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#inital_pt
a=np.array([6,5])
#shift_pt
b=np.array([-3,1])
#new_pt-->new Location point
new_point=a+b

plt.figure(figsize=(8,9))

#draw initial_pt
plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',color='red',scale=1)

#draw shift_point
plt.quiver(a[0],a[1],b[0],b[1],angles='xy',scale_units='xy',scale=1,color='blue')

#draw new_point
plt.quiver(0,0,new_point[0],new_point[1],angles='xy',scale_units='xy',scale=1,color='green')

from matplotlib.lines import Line2D
legend_elements=[
    Line2D([0],[0],color='red',lw=2,label='Initial(6,5)'),
    Line2D([0],[0],color='blue',lw=2,label='Shift(-3,1)'),
    Line2D([0],[0],color='green',lw=2,label='New Point(3,6)')
]

plt.xlim(-1,9)
plt.ylim(-1,9)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.title('Geometric Vector Addition')
plt.grid(True)
plt.legend(handles=legend_elements)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([5,5])
scaler=2.5
scaled=x*scaler

plt.figure(figsize=(10,10))
plt.quiver(0,0,x[0],x[1],angles='xy',scale_units='xy',scale=1,color='black',label='X([5.23,8.34])',width=0.005)
plt.quiver(0,0,scaled[0],scaled[1],angles='xy',scale_units='xy',scale=1,color='blue',label='Scaled([13.075 ,20.85 ])',width=0.005)

plt.xlim(0, 40)
plt.ylim(0, 25)
plt.grid(True)
plt.title('Effect of Scalar Multiplication (Scaling)')
plt.xlabel('Cups Required')
plt.ylabel('Cones Required')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.array([8,3])
b=np.array([-5,2])

n_pt=a+b
print("n_pt:",n_pt)

plt.figure(figsize=(7,8))

plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,label='a(8,3)',color='orange')
plt.quiver(a[0],a[1],b[0],b[1],angles='xy',scale_units='xy',scale=1,label='b(-5,2)',color='red')
plt.quiver(0,0,n_pt[0],n_pt[1],angles='xy',scale_units='xy',scale=1,label='n_point(3,5)',color='black')

plt.xlim(0,9)
plt.ylim(0,9)
plt.xlabel('Stocks')
plt.ylabel('Quantity')
plt.title('Geometric Co-Ordinate representation')
plt.grid(True,alpha=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
c=np.array([2.26,3.8])
scaler=4.3

scaled=c*scaler

print(scaled)
plt.figure(figsize=(8,8))
plt.quiver(0,0,c[0],c[1],angles='xy',scale_units='xy',scale=1,label='c(2.26,3.8)',color='green')
plt.quiver(0,0,scaled[0],scaled[1],angles='xy',scale_units='xy',scale=1,label='Scaled(9.718,16.34)',color='red')

plt.xlim(0,20)
plt.ylim(0,20)
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.title('Geometric Price & Quantity Fluctuation',fontweight='bold',fontsize=18)
plt.grid(True)
plt.legend()
plt.show()
```
