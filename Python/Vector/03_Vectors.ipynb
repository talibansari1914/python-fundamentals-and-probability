# Linear Independence & Dimensions
### 1.Implementaion of basis Vector

```python
import numpy as np

basis_1=np.array([1.,0.,0.])
basis_2=np.array([0.,1.,0.])
basis_3=np.array([0.,0.,1.])

scaled_basis_vector=5*basis_1+10*basis_2+15*basis_3
print("Basis Vector 1:",basis_1)
print("Basis Vector 2:",basis_2)
print("Basis Vector 3:",basis_3)
print("scaled basis vector :",scaled_basis_vector)
```

```python
import torch
basis_t1=torch.tensor([1,0,0],dtype=torch.float32)
basis_t2=torch.tensor([0,1,0],dtype=torch.float32)
basis_t3=torch.tensor([0,0,1],dtype=torch.float32)

scaled_vector_torch=2*basis_t1+4*basis_t2+6*basis_t3
print("Torch Basis Vector:",basis_t1.tolist())
print("Torch basis Vector:",basis_t2.tolist())
print("Torch basis Vector :",basis_t3.tolist())
print("Scaled Torch basis Vector:",scaled_vector_torch.tolist())
```

### Linear Independence of Basis

```python
import numpy as np
v_1=np.array([1,0])
v_2=np.array([0,1])
M_independent=np.array([v_1,v_2]).T
print("M_independent:",M_independent)

# Rank of matrix
Rank_independent=np.linalg.matrix_rank(M_independent)

v_3_redundant=np.array([1,1])

M_dependent=np.array([v_1,v_2,v_3_redundant]).T
print("M_Dependent:",M_dependent)
Rank_dependent=np.linalg.matrix_rank(M_dependent)

print("Rank Independent:",Rank_independent)
print("Rank Dependent:",Rank_dependent)
```

```python
import torch
v_1_t=torch.tensor([1.,0.,0.])
v_2_t=torch.tensor([0.,1.,0.])
v_3_t=torch.tensor([0.,0.,1.])
m_independent=torch.stack([v_1_t,v_2_t,v_3_t],dim=1)
print("Matrix Independent:",m_independent.tolist())

rank_independent_matrix=torch.linalg.matrix_rank(m_independent)
print("Rank of Independent Matrix:",rank_independent_matrix)
```

# Linear Independence & Redundency

```python
# example.1
import numpy as np
import matplotlib.pyplot as plt
v_1=np.array([2,3])
#v_2 is linearly indpendent on v_1 if v_1 is change then v_2 will also be changed
v_2=2*v_1

# ratio of v2/v1
v_3=v_2/v_1

print("V1:",v_1)
print("V2:",v_2)
print("V3:",v_3)

plt.figure(figsize=(8,8))
plt.quiver(0,0,v_1[0],v_1[1],angles='xy',scale_units='xy',scale=1,label='v_1(2,3)',color='orange')
plt.quiver(0,0,v_2[0],v_2[1],angles='xy',scale_units='xy',scale=1,label='v_2(3*[2,3])=v_2(4,6)',color='green')
plt.quiver(0,0,v_3[0],v_3[1],angles='xy',scale_units='xy',scale=1,label='Ratio(v_2/v_1)=v_3(2,2)',color='gray')

plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('X-axis',fontweight='bold',fontsize=12,color='black')
plt.ylabel('Y-axis',fontweight='bold',fontsize=12,color='black')
plt.title('Geometric Representation of Co-Ordinates',fontweight='bold',fontsize=16,color='green')
plt.grid(True,alpha=0.5)
plt.legend()
plt.show()
           
```

```python
import torch
t_1=torch.tensor([3,4])
t_2=3*t_1

#ratio
t_3=t_2/t_1

print("Tensor Data T1:",t_1.tolist())
print("Scaled Tensor Data T2:",t_2.tolist())
print("Ratio(shrinked data):",t_3.tolist())
```

```python
import torch
t1=torch.tensor([6,8])
t2=0.5*t1

m_dependent=torch.stack([t1,t2],dim=1)

# determinant
m_det=torch.det(m_dependent)

print("Tensor Data T1:",t1.tolist())
print("Scaled or Dependent Tensor Data:",t2.tolist())
print("Dependent Matrix:",m_dependent)
print("Determinant of dpendent Matrix:",m_det)
```

**Dependency check**

```python
import numpy as np
# vectors where all dependent on the z axis
a=np.array([1,0,0])
b=np.array([0,1,0])
c=np.array([1,1,0])

d,e,f=1,1,-1

result=d*a+e*b+f*c

print("Data a:",a)
print("Data b:",b)
print("Data c:",c)
print("Result:",result)
print("is all are dependent?:",np.allclose(result,np.zeros(3)))
```

**Dependent Vector**

```python
# example.2
import numpy as np
import matplotlib.pyplot as plt
# dependent vectors
v_A=np.array([1,2])
v_B=np.array([3,6]) # v_A*3

plt.figure(figsize=(8,8))
plt.quiver(0,0,v_A[0],v_A[1],angles='xy',scale_units='xy',scale=1,label='v_A(1,2)',color='red')
plt.quiver(0,0,v_B[0],v_B[1],angles='xy',scale_units='xy',scale=1,label='v_B(3,6)',color='green')

plt.xlim(-1,7)
plt.ylim(-1,7)
plt.xlabel('X-axis(v_A Co-Ordinates)',color='black',fontweight='bold',fontsize=10)
plt.ylabel('Y-axis(v_B Co-Ordinates)',color='black',fontweight='bold',fontsize=10)
plt.title('Geometric Representation',color='gray',fontweight='bold',fontsize=16)

plt.grid(True,alpha=0.5)
plt.legend()
plt.show()

```

```python
# example.3
import torch
v1_t=torch.tensor([1.,0.,0.])
v2_t=torch.tensor([0.,1.,0.])
v3_t=torch.tensor([0.,0.,1.])
v4_t=torch.tensor([3.,4.,0.]) # 3*v1_t + 4*v2_t
m_dependent=torch.stack([v1_t,v2_t,v3_t,v4_t],dim=1)

# Rank of dependent Matrix
rank_matrix=torch.linalg.matrix_rank(m_dependent)
print("Dependent Matrix:",m_dependent)
print("Rank of Dependent Matrix:",rank_matrix)
```
