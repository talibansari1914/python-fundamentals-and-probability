# Joint & Marginal Probability

```python
import pandas as pd
data={
    "Gender":["Boys","Boys","Girls","Girls"],
    "Result":["Pass","Fail","Pass","Fail"],
    "Count":[30,20,40,10]
}
df=pd.DataFrame(data)
print(df)
total=df["Count"].sum()

df["joint_prob"]=df["Count"]/total
print(df)
marginal_gender=df.groupby("Gender")["joint_prob"].sum()
marginal_result=df.groupby("Result")["joint_prob"].sum()
print("\nMarginal Probability (Gender):",marginal_gender)
print("\nMarginal Probability (Result):",marginal_result)

joint=df[(df["Gender"]=="Boys") & (df["Result"]=="Pass")]["joint_prob"].values[0]

cond=joint/marginal_gender["Boys"]
print("Joint Probability:",joint)
print("Conditional Probability:",cond)
```

```python
# Ek deck se card nikala gaya.
# Find:
# (a) P(King âˆ© Red)
# (b) P(King)
# (c) P(Red | King)
import pandas as pd

sample_space={
    "color":["R","R","B","B"],
    "type":["king","non_king","king","non_king"],
    "count":[2,24,2,24]
}
df=pd.DataFrame(sample_space)
total=df["count"].sum()
df["joint_pr"]=df["count"]/total

marginal_color=df.groupby("color")["joint_pr"].sum()
marginal_type=df.groupby("type")["joint_pr"].sum()

print("\nMarginal Probability (Color):",marginal_color)
print("\nMarginal Probability (Type):",marginal_type)

joint_p=df[(df["color"]=="R" ) & (df["type"]=="king")]["joint_pr"].values[0]
conditional=joint_p/marginal_type["king"]
print("\n Joint Probability:",joint_p)
print("\nConditional Probability:",conditional)
    
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

def visualize_bnb_concept():
    """
    Create a conceptual Bernouli Naive Bayes For spam detection,
    showing how word presence/absence "votes" for spam vs ham.
    """

    fig,(ax1,ax2)=plt.subplots(1,2,figsize=(16,7))

    plt.suptitle("Bernouli Naive Bayes:'How words 'Vote' for Spam or Ham",fontsize=20,weight="bold",y=1.0)

     # --- 1. Define Conceptual Word Likelihoods (The Model's Learned Knowledge) ---
    # These are simplified P(Word Present | Class) values

    word_data={
    "free":{"spam_prob":0.9,"ham_prob":0.05,"color":"red"},
    "lottery":{"spam_prob":0.8,"ham_prob":0.02,"color":"red"},
    "winner":{"spam_prob":0.7,"ham_prob":0.03,"color":"red"},
    "meeting":{"spam_prob":0.1,"ham_prob":0.9,"color":"green"},
    "report":{"spam_prob":0.05,"ham_prob":0.8,"color":"green"},
    "client":{"spam_prob":0.15,"ham_prob":0.75,"color":"green"},
    "urgent":{"spam_prob":0.6,"ham_prob":0.2,"color":"orange"} # natural weak
    }

    # --- 2. Simulate an Incoming Email ---
    # Example 1: Clear Spam
    # incoming_email_words = ['urgent', 'lottery', 'winner', 'free']

    # Example 2: Ambiguous Email
    incoming_email_words=["urgent","client","free","report"]

     # --- AX1: Word Likelihoods (The Model's Brain) ---
    ax1.set_title("1.The Model's Learnt knowledge(word 'Vote')",fontsize=16)
    ax1.set_xlim(-0.5,len(word_data)-0.5)
    ax1.set_ylim(-1,1)
    ax1.axhline(0,color="grey",linestyle="--",linewidth=0.8) # Natural Line

    # Create dummy scatter points for legend and labels
    ax1.scatter([],[],color="red",s=100,label="Strongly for SPAM(High P(Word | Spam))")
    ax1.scatter([],[],color="green",s=100,label="Stringly for HAM(High P(Word | Ham))")
    ax1.scatter([],[],color="orange",s=100,label="Weak/Natural Vote")

    word_x_pos=0
    annotations=[]

    for word,prob in word_data.items():
        spam_vote_strength=prob["spam_prob"]-0.5
        ham_vote_strength=prob["ham_prob"]-0.5

         # Combine into a single "net vote" for visualization
        # Positive = favors spam, Negative = favors ham
        net_vote=spam_vote_strength - ham_vote_strength

        color=prob["color"]
        ax1.scatter(word_x_pos,net_vote,color=color,s=200,zorder=5)

        # Annotate word
        text_color="red" if net_vote>0.2 else("green" if net_vote < -0.2 else "black")
        annotations.append(ax1.text(word_x_pos, net_vote + 0.1, word.upper(), ha='center', va='bottom', fontsize=12, color=text_color, weight='bold'))

        # Add tiny lines to indicate direction of vote
        ax1.annotate('', xy=(word_x_pos, net_vote * 0.5), xytext=(word_x_pos, 0),
                     arrowprops=dict(facecolor=color, shrink=0.05, width=1, headwidth=6, headlength=8),
                     zorder=4)
        word_x_pos += 1

        ax1.set_xticks([]) # Hide x-axis ticks
        ax1.set_yticks([-0.8, -0.4, 0, 0.4, 0.8])
        ax1.set_yticklabels(['Strong HAM Vote', 'Moderate HAM Vote', 'Neutral', 'Moderate SPAM Vote', 'Strong SPAM Vote'])
        ax1.set_ylabel('Net "Vote" Strength', fontsize=12)
        ax1.legend(loc='upper left', fontsize=10, frameon=True, edgecolor='black')
        ax1.grid(axis='y', linestyle='--', alpha=0.7)

         # Adjust annotation positions to avoid overlap
        from adjustText import adjust_text
        adjust_text(annotations, ax=ax1, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

       # --- AX2: Email Classification (The "Voting" Process for a New Email) ---
        ax2.set_title(f"2. Classifying New Email: '{' '.join(incoming_email_words).upper()}'", fontsize=16)
        ax2.set_xlim(-0.5, len(incoming_email_words) + 0.5)
        ax2.set_ylim(-1.5, 1.5)
        ax2.axhline(0, color='gray', linewidth=0.8, linestyle='--') # Neutral line  

        ax2.text(0.5, 1.3, "Overall 'Vote' for SPAM", ha='center', va='bottom', fontsize=12, color='red')
        ax2.text(0.5, -1.3, "Overall 'Vote' for HAM", ha='center', va='top', fontsize=12, color='green')

        current_spam_score = np.log(0.5) # Initial neutral prior (log scale for multiplication)
        current_ham_score = np.log(0.5)  # Initial neutral prior

        line_segments_spam = []
        line_segments_ham = []

        current_x = 0
        final_score_spam = 0 # for visual only
        final_score_ham = 0 # for visual only

        for i, word in enumerate(incoming_email_words):
            if word in word_data:
                spam_prob_present = word_data[word]['spam_prob']
                ham_prob_present = word_data[word]['ham_prob']

            # For words NOT present, we would use (1 - P(Word Present | Class))
            # But for simplicity, we focus on words *present* in this visual

            # Update scores (simulating multiplication by adding log-probabilities)
                current_spam_score += np.log(spam_prob_present)
                current_ham_score += np.log(ham_prob_present)

            # --- Visualization of Individual Word's Impact ---

            # The word itself

                ax2.text(current_x, 0, word.upper(), ha='center', va='center', fontsize=13, weight='bold', bbox=dict(facecolor='white', alpha=0.9, edgecolor='black', boxstyle='round,pad=0.3'))

            # Arrow towards SPAM score
                line_segments_spam.append([current_x, 0])
                line_segments_spam.append([current_x, spam_prob_present])

                ax2.annotate(f"{spam_prob_present:.1f}", xy=(current_x, spam_prob_present), xytext=(current_x, spam_prob_present + 0.1),
                            ha='center', va='bottom', fontsize=9, color='red', weight='bold')

            # Arrow towards HAM score
                line_segments_ham.append([current_x, 0])
                line_segments_ham.append([current_x, -ham_prob_present]) # Plotting ham votes downwards


                line_segments_ham.append([current_x, 0])

                line_segments_ham.append([current_x, -ham_prob_present]) # Plotting ham votes downwards

                ax2.annotate(f"{ham_prob_present:.1f}", xy=(current_x, -ham_prob_present), xytext=(current_x, -ham_prob_present - 0.1),
                            ha='center', va='top', fontsize=9, color='green', weight='bold')

            # Add a vertical separator
                if i < len(incoming_email_words) - 1:
                    ax2.axvline(current_x + 0.5, color='gray', linestyle=':', alpha=0.7)

                final_score_spam += spam_prob_present # For visual line only
                final_score_ham += ham_prob_present # For visual line only
                current_x += 1

            else: # Word not in our model's vocabulary
                    ax2.text(current_x, 0, f"{word.upper()}\n(Unknown)", ha='center', va='center', fontsize=11, color='gray', bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray', boxstyle='round,pad=0.2'))
                    current_x += 1

            # --- Final Result ---
            # Approximate final decision based on sums of probs (conceptual, not actual Bayes)

        final_decision_y = 0
        final_decision_text = ""

        if final_score_spam > final_score_ham:
            final_decision_y = 1.0
            final_decision_text = "PREDICTED: SPAM!"
            final_decision_color = 'red'
        elif final_score_ham > final_score_spam:
            final_decision_y = -1.0
            final_decision_text = "PREDICTED: HAM!"
            final_decision_color = 'green'
        else:
           final_decision_y = 0
           final_decision_text = "PREDICTED: NEUTRAL!"
           final_decision_color = 'gray'

        ax2.plot(np.linspace(0, current_x - 1, len(line_segments_spam)//2), [0]* (len(line_segments_spam)//2), 'ko', markersize=6, label="Word Present", zorder=5) # Mark word positions

        ax2.set_xticks([])
        ax2.set_yticks([-1, 0, 1])
        ax2.set_yticklabels(['Favors HAM', 'Neutral', 'Favors SPAM'])
        ax2.set_ylabel('Individual Word Influence', fontsize=12)

    # Plot cumulative "vote" for visualization
        spam_trace_y = [0] * len(incoming_email_words)
        ham_trace_y = [0] * len(incoming_email_words)

        current_spam_trace = 0
        current_ham_trace = 0

        for idx, word in enumerate(incoming_email_words):
            if word in word_data:
                current_spam_trace += word_data[word]['spam_prob']
                current_ham_trace += word_data[word]['ham_prob']

            spam_trace_y[idx] = current_spam_trace
            ham_trace_y[idx] = current_ham_trace

        # Visualizing the final cumulative "vote" or push
        ax2.text(current_x / 2, final_decision_y * 0.7, final_decision_text,
                 fontsize=20, weight='bold', color=final_decision_color,
                 ha='center', va='center',
                 bbox=dict(facecolor='white', alpha=0.9, edgecolor=final_decision_color, boxstyle='round,pad=0.5'))

        ax2.axvline(x=-0.5, color='black', linestyle='-', linewidth=2) # Left border
        ax2.axvline(x=len(incoming_email_words)-0.5, color='black', linestyle='-', linewidth=2) # Right border
        ax2.axhline(y=1.5, color='black', linestyle='-', linewidth=2) # Top border
        ax2.axhline(y=-1.5, color='black', linestyle='-', linewidth=2) # Bottom border

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

# Run the visualization
    visualize_bnb_concept()

```
