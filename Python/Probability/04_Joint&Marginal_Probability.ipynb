## Joint Probability

**Where 2 or more events happens together**<br>
**Means-** toss of coin and roll of dice can happen together<br>
$$P(A and B)=P(A∩B)$$
$$P(A and B)= P(A)*P(B)$$

```python
#  Example 1:
# A coin is tossed and a die is rolled.
# Find probability of getting Head and number 4.
import random

# when a coin tossed the possible outcomes=["head","tail"]
P_H=1/2

# When a dice is rolled the possible outcomes=[1,2,3,4,5,6]
P_4=1/6

# Joint Probability
P_H_4=P_H*P_4

print("Probability of Head \n P(H) ~",P_H)
print("Probability of 4 \n P(4) ~",P_4)
print("Joint Probability ~",P_H_4)
    
```

```python
# Example 2 (Dependent Events):
# Deck of 52 cards, 2 draws without replacement
# Probability of drawing two Kings.

P_1K=4/52 # there are 52 cards and there are 1 king in each of set of cards
P_2K=3/51 # 1 king we have taken and we can not replace it so 4 became 3 and 52 became 51

# joint Probability
P_1K_2K=P_1K*P_2K
# we are using formula here as we know that P(A/B)=P(A and B)/P(B)
                                        #   P(A/B)*P(B)=P(A and B)

print(" 1 king draw \n P_1K ~",P_1K)
print("2 king draw \n P_2K ~",P_2K)
print("Joint Probability (P_1K_2K) ~",P_1K_2K)
```
