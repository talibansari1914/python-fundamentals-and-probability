## 1.. What is Law of Total Probability?

- **Law of Total Probability (LTP) tells us how to find the probability of an event (A)**<br>
 **when that event can happen through several different scenarios (Bâ‚, Bâ‚‚, Bâ‚ƒ, ...)**<br>
- **Means-** When multiple conditions are given then we have to find the probability<br>
- **If B1,B2,B3,...,Bn are mutually exclusive and exhaustive events (cover all possibilities), then:**<br>
$$P(A)=P(Aâˆ£B1â€‹)P(B1â€‹)+P(Aâˆ£B2â€‹)P(B2â€‹)+...+P(Aâˆ£Bnâ€‹)P(Bnâ€‹)$$

```python
#  Example 1:
# A company has 3 factories producing bulbs:
# Factory Fâ‚ â†’ 30% of bulbs
# Factory Fâ‚‚ â†’ 50% of bulbs
# Factory Fâ‚ƒ â†’ 20% of bulbs
# Each factoryâ€™s defective rate is:
# Fâ‚ â†’ 1% defective
# Fâ‚‚ â†’ 3% defective
# Fâ‚ƒ â†’ 2% defective
# Now find: Probability that a bulb (chosen randomly) is defective.

P_F1=0.3
P_F2=0.5
P_F3=0.2

# probability of defective
P_DF1=0.01
P_DF2=0.03
P_DF3=0.02

# PROBABILITY OF A BULB THAT IS DEFECTIVE :
P_B=P_F1*P_DF1+P_F2*P_DF2+P_F3*P_DF3

print("Probability of chosing the defective Bulb is ~",P_B)
```

```python
# Example 2: Rain Prediction 
# 40% days are cloudy â†’ P(C) = 0.4
# 60% days are clear â†’ P(Câ€™) = 0.6
# If cloudy â†’ 70% chance of rain â†’ P(R|C)=0.7
# If clear â†’ 10% chance of rain â†’ P(R|Câ€™)=0.1
# Find P(Rain).
# ğ‘ƒ(ğ‘…)=ğ‘ƒ(ğ‘…âˆ£ğ¶)ğ‘ƒ(ğ¶)+ğ‘ƒ(ğ‘…âˆ£ğ¶â€²)ğ‘ƒğ¶â€²)P(R)=P(Râˆ£C)P(C)+P(Râˆ£Câ€²)P(Câ€²)

P_C=0.4 # probability of days cloudy
P_not_C=0.6 # Probability of days not cloudy
P_R_C=0.7 # probability of chances of rain
P_not_R_C=0.1 # probability of not rain

P_R=P_C*P_R_C+P_not_C*P_not_R_C # total probability of rain with all the conditions
print("Total Probability of Rain ~",P_R)
```

```python
# Example 3: Student Passing in Exam ğŸ“
# A student can prepare from 3 books:
# Book1 (40% students use) â†’ Pass rate 90%
# Book2 (35% students use) â†’ Pass rate 80%
# Book3 (25% students use) â†’ Pass rate 70%
# Find overall probability a random student passes.

P_B1=0.9 # Book 1 pass rate
P_B2=0.8 # Book 2 pass rate
P_B3=0.7 # Book 3 pass rate

P_S1=0.4 # Boook 1 students used
P_S2=0.35 # Book 2 students used 
P_S3=0.25 # Book 3 students used


P_stud_Pass=P_B1*P_S1+P_B2*P_S2+P_B3*P_S3
print("Random student passes probability ~",P_stud_Pass)
```

```python
# Example 4:Three machines: 40%, 35%, 25% production; defect rates 5%, 6%, 7%

P_M1=0.4 # Machine 1 Production
P_M2=0.35 # Machine 2 Production
P_M3=0.25 # Machine 3 Production

P_D1=0.05 # Machine 1 defect rate probability
P_D2=0.06 # Machine 2 defect rate probability
P_D3=0.07 # Machine 3 defect rate probability
# so the Total Probability is:
P_M=P_M1*P_D1+P_M2*P_D2+P_M3*P_D3
print("Machine Production with defect rate ~",P_M)
```
