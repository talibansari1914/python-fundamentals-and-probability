## 1.What is conditional probability ?

**we find here P(A) when P(B) is already occured**

$$P(A∣B)=P(A and B)/P(B)$$​

## Examples

```python
# 1.problem.we draw 2 cards without replacement from a deck of 52 cards.
# Find probability that both are Aces.
total_cards=52
total_aces=4 # 1 in each 13 set of cards

first_ace=total_aces/total_cards
# without replacement
second_ace=(total_aces-1)/(total_cards-1)

# Probability of both the aces
P_F_S=first_ace*second_ace

print("Probability of first aces ~",first_ace)
print("Probability of second aces ~",second_ace)
print(" Probability of both the aces ~",P_F_S)
```

```python
# 2.problem. Conditional Probability for Dependent Events
# Example:
# 40% students play football → P(F) = 0.4
# 25% play cricket → P(C) = 0.25
# 15% play both → P(F and C) = 0.15
# Now find:
# P(F|C) = ?
# P(C|F) = ?

P_F=0.4
P_C=0.25
P_F_and_C=0.15 #P(A and B)

# P(F/C)=P(F and C)/P(C)

P_F_C=P_F_and_C/P_C

#P(C/F)=P(F and C)/P(F)
P_C_F=P_F_and_C/P_F

print("P_F_C ~",P_F_C)
print("P_C_F ~",P_C_F)
```

## Independence Check Using Conditional Probability

**If A and B are independent,**
**then:**
$$P(A∣B)=P(A)$$

```python
# Tossing two coins:
# P(A) = first coin → Head = 0.5
# P(B) = second coin → Head = 0.5
# P(A and B) = 0.25

P_A=0.5
P_B=0.5
P_A_and_B=0.25

# P(A/B)=P(A and B)/P(B)

P_A_B_Given=P_A_and_B/P_B

print("P_A ~",P_A)
print("P_A_B_Given ~",P_A_B_Given)
```

```python
# problem.rolling two dice
# 1.first count ==4
# 2.sum(both dice)>8

import random
trial=1000
dice1=0
count_both=0

for _ in range(trial):
    d1=random.randint(1,6)
    d2=random.randint(1,6)
    if d1==4:
        dice1+=1
        if d1+d2>8:
            count_both+=1
P_Sum_Given_4=count_both/dice1
print("P_Sum_Given_4 ~",P_Sum_Given_4)
    
```

```python
# problem.rolling two dice
# 1.first dice=3
# 2.second dice=3 find the probability when 3 of second dice is already occured

import random
trial2=1000
count_d1=0
count_d2=0
for _ in range(trial2):
    dice1=random.randint(1,6)
    dice2=random.randint(1,6)
    if dice2==3:
        count_d2+=1
        if dice1==3:
            count_d1+=1
P_d1_3_d2_3_given=count_d1/count_d2
print("Probability of occuring 3 in dice 1 when in dice 2 ,3 is already occured\n P_d1_3_d2_3_given ~",P_d1_3_d2_3_given)
```
