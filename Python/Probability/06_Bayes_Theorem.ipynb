## 1. What is Bayesâ€™ Theorem?

- **Bayesâ€™ Theorem helps you find reverse probability â€”**
**yaani agar tumhe â€œA given Bâ€ chahiye, lekin tumhe â€œB given Aâ€ pata hai.**<br>
**Formula:**
$$P(Aâˆ£B)=P(Bâˆ£A)Ã—P(A)/P(B)â€‹$$

```python
Derivation:

  we know conditional Probability

          P(A|B)=P(A and B)/P(B)  --------(1)    # here as we know that 
                                                 # 1. probability of A when B is Given

          P(B|A)=P(A and B)/P(A)----------(2)   # here 1.probability of B when A is Given 

 # so from equation 1 and 2 we can say that
       P(A and B)=P(A|B)*P(B)
       P(A and B)=P(B|A)*P(A)
 # so we can do equal to both the equations
        P(A|B)*P(B)=P(B|A)*P(A)
# then
        P(A|B)=( P(B|A)*P(A) ) / P(B)
                  # This is called Bayes Theorem
```

```python
#  Example 1:A disease affects 1% of the population.
# A test for it is:
# 99% accurate when the person is sick â†’ P(Pos | Disease) = 0.99
# 95% accurate when the person is healthy â†’ P(Neg | No Disease) = 0.95
# Now find:
# Probability that a person actually has the disease given they test positive.

P_D=0.01 # as only 1% people got affected
P_not_D=1-P_D # saved people
P_P_D_given=0.99 # P(POs/Disease)
P_P_not_D_given=0.05 # P(Neg/No disease)

# so the Total Probability is:
P_T=P_P_D_given*P_D+P_not_D*P_P_not_D_given

# Probability that a person actually has the disease given they test positive.
P_D_P_given=(P_P_D_given*P_D)/P_T
print(" Probability that a person actually has the disease given they test positive ~",P_D_P_given)

```

```python
# Example 2.There are 3 factories:
# | Factory | P(F) | P(Defective | F) |
# |----------|------|---------------|
# | F1 | 0.3 | 0.01 |
# | F2 | 0.5 | 0.03 |
# | F3 | 0.2 | 0.02 |
# Now find:
# Probability that a defective bulb came from Factory 2 (F2).

P_F1=0.3
P_F2=0.5
P_F3=0.2

P_D_F1=0.01 # P(Defective|F)
P_D_F2=0.03
P_D_F3=0.02

# total probability 
P_T=P_F1*P_D_F1+P_F2*P_D_F2+P_F3*P_D_F3

# Defecetive bulb from factory 2
P_F_D=(P_D_F2*P_F2)/P_T

print(f"Probability of Defective Bulb From Factory 2 is ~{P_F_D:.3f}")
```

```python
# Problem 1 â€” Medical Test (Easy)
# A disease affects 1% of people.
# A test for it gives a true positive 98% of the time and a false positive 3% of the time.
# Find:
# ğŸ‘‰ Probability that a person actually has the disease given that the test result is positive.
# ğŸ’¡ Given:
# P(D) = 0.01
# P(Â¬D) = 0.99
# P(Pos | D) = 0.98
# P(Pos | Â¬D) = 0.03
# Find:
# ğ‘ƒ(ğ·âˆ£ğ‘ƒğ‘œğ‘ )=?P(Dâˆ£Pos)=?

P_D_effect=0.01 # disease effect probability
P_D_not_Effect=1-P_D_effect # not disease effect probability
P_Pos_D_given=0.98 # P(pos|D)
P_Pos_not_D=0.03 # False positive probability

# Total Probability of coming Positive
P_T=P_D_effect*P_Pos_D_given+P_D_not_Effect*P_Pos_not_D

# Probability of getting disease when report is positive
P_D_pos=(P_Pos_D_given*P_D_effect)/P_T

print(f" Probability of getting disease when report is positive ~{P_D_pos:.3f}")
```

```python
# Problem 2 â€” Coin Toss (Medium)
# There are two coins:
# Coin A: Fair coin (P(Head) = 0.5)
# Coin B: Biased coin (P(Head) = 0.75)
# You randomly pick one coin and toss it.
# You get a Head.
# Find:
# ğŸ‘‰ Probability that you picked the biased coin.
# ğŸ’¡ Given:
# P(A) = P(B) = 0.5
# P(H | A) = 0.5
# P(H | B) = 0.75
# Find:
# ğ‘ƒ(ğµâˆ£ğ»)=?P(Bâˆ£H)=?

P_A=P_B=0.5
P_H_A_given=0.5
P_H_B_given=0.75

P_T=P_A*P_H_A_given+P_B*P_H_B_given

P_B_H=(P_H_B_given*P_B)/P_T
print(f"Probability of biased coin getting head when it is tossed ~{P_B_H:.3f}")
```

```python
# Problem 3 â€” Factory Defects (Medium)
# Three factories Fâ‚, Fâ‚‚, and Fâ‚ƒ produce 20%, 50%, and 30% of the companyâ€™s output.
# Their defective rates are 1%, 4%, and 3% respectively.
# A randomly chosen product is defective.
# Find:
# ğŸ‘‰ Probability it came from Factory Fâ‚‚.
# ğŸ’¡ Given:
# P(Fâ‚) = 0.2, P(Fâ‚‚) = 0.5, P(Fâ‚ƒ) = 0.3
# P(D | Fâ‚) = 0.01, P(D | Fâ‚‚) = 0.04, P(D | Fâ‚ƒ) = 0.03
# Find:ğ‘ƒ(ğ¹2âˆ£ğ·)=?P(F2âˆ£D)=?

P_F1=0.2
P_F2=0.5
P_F3=0.3

# defectives of factories
P_D_F1=0.01
P_D_F2=0.04
P_D_F3=0.03

# By the Law of Total Probability finding total defect
P_T=P_F1*P_D_F1+P_F2*P_D_F2+P_F3*P_D_F3

# P(f2|D)
P_F2_D=(P_D_F2*P_F2)/P_T
print(f"Random choosen Defective ~{P_F2_D:.3f}")
```

```python
# Problem 4 â€” Email Spam (Mediumâ€“Hard)
# An email is spam 40% of the time.
# If an email is spam, the word â€œWINâ€ appears 70% of the time.
# If itâ€™s not spam, the word â€œWINâ€ appears 10% of the time.
# You see an email containing â€œWINâ€.
# Find:
# ğŸ‘‰ Probability that it is spam.
# ğŸ’¡ Given:
# P(S) = 0.4, P(Â¬S) = 0.6
# P(W | S) = 0.7, P(W | Â¬S) = 0.1
# Find:ğ‘ƒ(ğ‘†âˆ£ğ‘Š)=?P(Sâˆ£W)=?

P_S=0.4
P_not_S=1-P_S
P_W_S=0.7
P_W_not_S=0.1

# By the Law of Total Probability
P_T=P_S*P_W_S+P_not_S*P_W_not_S

# P(S|W)
P_S_W=(P_W_S*P_S)/P_T
print(f"Probability when message have win it is spam or not ~{P_S_W:.3f}")
```
