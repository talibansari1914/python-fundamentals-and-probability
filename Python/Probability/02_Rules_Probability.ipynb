## Sum Rule
**For Mutually Exclusive Event**<br>
**Means-** Two Event Happens independently they are not overlapping to eachother<br>
$$P(A or B)=P(A)+P(B)$$

## example

```python
import random
trial1=10000
# Probability of Head which is P_H
head=0
for _ in range(trial1):
    if random.choice(["Head","Tail"])=="Head":
        head+=1
P_H=head/trial1

# Probability of Tail which is P_T
tail=0
for _ in range(trial1):
    if random.choice(["Head","Tail"])=="Tail":
        tail+=1
P_T=tail/trial1

# Mutually Exclusive probability
P_H_or_T=P_H+P_T

print("P_H ~",P_H)
print("P_T ~",P_T)
print("P_H_or_T ~",P_H_or_T)
```

**Non - Mutually Exclusive Event**<br>
**Means-**Both A and B Event Are dependent or overlapping on each other<br>
$$P(A or B)=P(A)+P(B)‚àíP(A and B)$$

```python
# Problem- There are 200 student in the school XYZ .Where 50 students used to play cricket and 70 used to play football
# but there are only 20 students who used to play both the games.find the probability of P(A),P(B),(A and B),P(A or B).


total_students=200
cricket_stud=50
football_stud=70
cricket_football=20

# Probability of cricket Playing student
P_C=cricket_stud/total_students

# Probability of football playing students
P_F=football_stud/total_students

# Probability of students who play cricket and football both
P_C_F=cricket_football/total_students

# probability of total students who play cricket and football
P_C_or_F=P_C+P_F-P_C_F

print("---Probability of cricket playing students---\n P_C ~",P_C)
print("---Probability of footlball playing students--\n P_F ~",P_F)
print("---Probability of Both Cricket and Football Playing students---\n P_C_F ~",P_C_F)
print("---Probability of Total students who play cricket and football--- \n P_C_or_F ~",P_C_or_F)
```

## Multiplication Rule
***1.first and second event both are independent event**
$$P(A and B)=P(A)√óP(B)$$

```python
# problem- if we toss 2 coins for 10 times what is the probability of the head
import random
toss=10
# for coin1
head1=0
for _ in range(toss):
    if random.choice(["Head","Tail"])=="Head":
        head1+=1
P_H1=head1/toss

# for coin2
head2=0
for _ in range(toss):
    if random.choice(["Head","Tail"])=="Head":
        head2+=1
P_H2=head2/toss

# Total probability of P_H1 and P_H2 because they are independent event
P_H1_and_H2=P_H1*P_H2

print("First coin Head Probability\n P_H1 ~",P_H1)
print("Second coin head Probability\n P_H2 ~",P_H2)
print("Total Probability of P_H1_and_H2 ~",P_H1_and_H2)
```

## Complementory Rule
$$P(A‚Ä≤)=1‚àíP(A)$$

```python
# problem- if dice rolled 1000 times then what is the probability of not coming six
import random
roll=1000

six=0
for _ in range(roll):
    if random.choice([1,2,3,4,5,6])==6:
        six+=1
P_six=six/roll

# probability of not coming 6
P_not_six=1-P_six

print("========Probability========")
print("P_six ~",P_six)
print("P_not_six ~",P_not_six)
```

## 5. Conditional Probability Rule

**Rule:**
$$P(A‚à£B)=P(A and B)/P(B)$$


```python
# Problem:
# In a class of 100 students:
# 40 students study Mathematics (Event A)
# 50 students study Science (Event B)
# 20 students study both Mathematics and Science (Event A ‚à© B)
# Find the probability that a student studies Mathematics given that the student studies Science, i.e. 
# ùëÉ(ùê¥‚à£ùêµ)P.

total_stud=100
Math_stud=40
Science_stud=50
Math_Science=20

# Probability of Mathematics students
P_M=Math_stud/total_stud

# Probability of Science Students
P_S=Science_stud/total_stud

# Probability of Both who study science and math
P_M_and_S=Math_Science/total_stud

# probability that a student studies Mathematics given that the student studies Science
P_M_S=P_M_and_S/P_S

print("========Conditional Probability========\n")
print("---Probability of Student who studies math---\n P_M ~",P_M)
print("---Probability of Student who studies science---\n P_S ~",P_S)
print("---Probability of Student who studies both---\n P_M_and_S ~",P_M_and_S)
print("---probability that a student studies Mathematics given that the student studies Science---\n P_M_S ~",P_M_S)
```
