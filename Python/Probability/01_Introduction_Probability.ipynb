### Examples

#### 1.Probability of getting Head

```python
import random

# toss outcomes
outcomes=["Head","Tail"]
random=random.choice(outcomes)
print("---Random Outcomes of Coin when it Tossed---\n",random)
```

```python
import random
trials=1000
head=0
for _ in range(trials):
    if random.choice(["Head","Tail"])=="Head":
        head+=1
prob=head/trials
print("Pobabilit of Head is ~",prob)
```

#### 2.Probability of getting Tail


```python
import random
outcomes_2=["Tail","Head"]
random_2=random.choice(outcomes_2)
print("---Random outcomes of Coin When it tossed---\n",random_2)
```

```python
import random
trial=10000
tail=0
for _ in range(trial):
    if random.choice(["Tail","Head"])=="Tail":
        tail+=1
prob2=tail/trial
print("Probability of Tail ~",prob2)
```
