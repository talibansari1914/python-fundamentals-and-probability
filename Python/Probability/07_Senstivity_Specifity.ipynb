## 1. Sensitivity (True Positive Rate)

**Sensitivity = Probability(test positive | disease present)**<br>
**Meaning-Jab kisi ko disease hai, tab test usko correctly positive detect kare — uska chance.**<br>
**Formula:**<br>
$$Sensitivity=True Positives / False Negatives+True Positives$$
	​


```python
# Example:
# 100 logon me se 40 logon ko disease hai.
# Un 40 me se 36 test me positive aaye.
total_case=40
true_pos=36
false_negative=4

Sensitivity=true_pos/(true_pos+false_negative)

print(f"Senstivity Probability ~{Sensitivity:.2f}")
```

## 2.Specificity(True Negative)

**Specificity = Probability(test negative | disease absent)**<br>
**Meaning-Jab kisi ko disease nahi hai, tab test correctly negative detect kare — uska chance.**<br>
**Formula:**
$$Specificity=True Negatives / False Positives+True Negatives​$$
