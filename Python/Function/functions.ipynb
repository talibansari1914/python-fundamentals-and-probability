
8.Write a function to check whether a given string or number is a palindrome.

```python
s=input()
def check_palindrome(s):
    if s.isdigit():
        n=int(s)
        pali=n
        rev=0
        while n>0:
            digit=n%10
            rev=rev*10+digit
            n=n//10
        if rev==pali:
            print("s is palindrome")
        else:
            print("it is not a palindrome")
    else:
        char=''
        for j in s:
            char=j+char
        if char==s:
            print("it is a palindrome")
        else:
            print("it is not a palindrome")
            
check_palindrome(s)
```
